```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

def dayname(index):
    if index >= 0 and index < len(days):
        previous_day = days[(index - 1) % len(days)]
        present_day = days[index]
        next_day = days[(index + 1) % len(days)]
        return (previous_day, present_day, next_day)
    else:
        return None

index = 3
result = dayname(index)
if result:
    print(result)
else:
    print("Invalid index")
```
