```python
path1 = (
    "http://e-smartdata.teachable.com/p/"
    "sciezka-data-scientist-machine-learning-engineer"
)
path2 = (
    "http://e-smartdata.teachable.com/p/"
    "sciezka-data-scientist-machine-learning-engineer"
)
path3 = (
    "http://e-smartdata.teachable.com/p/"
    "sciezka-bi-analyst-data-analyst"
)

# Function to find the index of the word "scientist" in a given path
def scientist_index(path):
    index = path.find("scientist")
    return index

# Find the index of "scientist" in each path and print the result
index1 = scientist_index(path1)
index2 = scientist_index(path2)
index3 = scientist_index(path3)

print("Index in path1: ", index1 if index1 != -1 else -1)
print("Index in path2: ", index2 if index2 != -1 else -1)
print("Index in path3: ", index3 if index3 != -1 else -1)
```
