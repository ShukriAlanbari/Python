```python
rounds = 10         # Number of rounds to be played
line_length = 60    # Define the fixed line length

def get_player():
    '''Asks for input of two player names, returns them as a tuple.'''
    # Prompt the user to enter the first players name.
    p1_name = input("Enter the first players name: ")
    # Prompt the user to enter the second players name.
    p2_name = input("Enter the second players name: ")
    # Returning a tuple that contains both players name.
    return (p1_name, p2_name)

def welcome_to_game(name1, name2):
    '''Welcomes the players with their given names.'''

    print('=' * line_length + '\n' +
          'Welcome {} and {} to the game!'.format(name1,
                                                  name2).center(line_length)
          + '\n' + 'The game has now started. May the best player '
          'win!'.center(line_length) + '\n' +'=' * line_length)

def determine_winner(player_list: list = list(), p1_choice:str = '0',
                     p2_choice:str = '0', scores_list: list = list()) -> list :
    '''Function to determine winner between
    two players in rock, paper, scissors.'''

    # List of sets where the player wins
    winning_combinations = [('0', '2'), ('1', '0'), ('2', '1')]
    if p1_choice == p2_choice:  # Check if tied
        print(f"It's a draw. Current score: {scores_list}")
        return scores_list
    # Check if player winns by comparing the choices to winning_combinations
    elif (p1_choice, p2_choice) in winning_combinations:
        scores_list[0] += 1
        print(f"{player_list[0]} wins. Current score: {scores_list}")
        return scores_list
    else:   # In not tied or player win -> computer winns
        scores_list[1] += 1
        print(f"{player_list[1]} wins. Current score: {scores_list}")
        return scores_list

def game_logic(player_list: list = list()):
    '''Main game logic to handle rounds and scoring.'''

    scores = [0, 0]    # Initialize scores
    for i_round in range(rounds):
        print('\n' +'=' * line_length + '\n' +
              'Round {} of {}'.format(i_round+1, rounds).center(line_length)+
              '\n' + '=' * line_length + '\n')
        # Players make their choices
        p1_choice = input(f'{player_list[0]} pick 0 (rock) '
                          f'1 (paper) or 2 (scissors): ')
        p2_choice = input(f'{player_list[1]} pick 0 (rock) '
                          f'1 (paper) or 2 (scissors): ')

        # Default to '0' if invalid choice
        if p1_choice not in ['0', '1', '2']:
            p1_choice = '0'
        if p2_choice not in ['0', '1', '2']:
            p2_choice = '0'

        # Determine the winner of the round
        scores = determine_winner(player_list, p1_choice, p2_choice, scores)

    # Printing the end of the game message with final score and winner (or draw)
    print('\n' +'=' * line_length + '\n' +
      'END OF GAME'.center(line_length) +
      '\n' + '=' * line_length+ '\n')
    msg_score = f"Final score {scores}"
    msg_winner = \
    f"Winner is {player_list[0]}" if scores[0] > scores[1] else \
    f"Winner is {player_list[1]}" if scores[0] < scores[1] else \
    f"It's a draw between: {player_list[0]} and {player_list[1]}"
    print(f"{msg_score} {msg_winner}")

def main():
    '''
    The main function to orchestrate the Rock, Paper, Scissors game.

    This function prompts the players to enter their names,
    welcomes them to the game, and manages the overall game
    logic, including the number of rounds and scoring.

    Usage:
    - Call this function to start the Rock, Paper, Scissors game.
    '''
    player_names = get_player()
    welcome_to_game(player_names[0],player_names[1])
    game_logic(player_names)

main()  # Calling the function
```
